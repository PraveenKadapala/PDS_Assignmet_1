```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# importing studentsPerformance data
studentsPerformance_raw_data = pd.read_csv('/content/students_performance/data_raw/StudentsPerformance.csv')

# Changing the categorical values into numerical class values

studentsPerformance_raw_data['gender'].replace(['female', 'male'], [0, 1], inplace=True)
studentsPerformance_raw_data['test preparation course'].replace(['none', 'completed'], [0, 1], inplace=True)

#dropping the rows with na
studentsPerformance_raw_data = studentsPerformance_raw_data.dropna()

studentPerformance_clean_data = studentsPerformance_raw_data

# Analysis 1: Box plot of math score distribution by lunch type
plt.figure(figsize=(8, 6))
sns.boxplot(x='lunch', y='math score', data=studentPerformance_clean_data, hue='lunch', dodge=False)
plt.title('Math Score Distribution by Lunch Type')
plt.xlabel('Lunch Type')
plt.ylabel('Math Score')
plt.savefig('/content/students_performance/results/mathScore_lunchType.png')
plt.show()
```

```python
# Analysis 2: Bar plot of race/ethnicity distribution

plt.figure(figsize=(8, 5))
sns.countplot(x='race/ethnicity', hue='race/ethnicity', data=studentPerformance_clean_data, palette='Set1', legend=False)
plt.title('Race/Ethnicity Distribution')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Count')
plt.savefig('/content/students_performance/results/Race_Ethnicity_distribution.png')
plt.show()
```

```python
# Analysis 3: Box plots for math score, reading score, and writing score by gender

plt.figure(figsize=(14, 5))

plt.subplot(1, 3, 1)
sns.boxplot(x='gender', y='math score', data=studentPerformance_clean_data, hue='gender')
plt.title('Math Score by Gender')

plt.subplot(1, 3, 2)
sns.boxplot(x='gender', y='reading score', data=studentPerformance_clean_data, hue='gender')
plt.title('Reading Score by Gender')

plt.subplot(1, 3, 3)
sns.boxplot(x='gender', y='writing score', data=studentPerformance_clean_data, hue='gender')
plt.title('Writing Score by Gender')

plt.savefig('/content/students_performance/results/studentScores_gender.png')
plt.show()
```

```python
# Analysis 4: Histogram for showing the distribution of Reading Scores

plt.figure(figsize=(8, 5))
sns.histplot(studentPerformance_clean_data['reading score'])
plt.title('Distribution of Reading Scores')
plt.xlabel('Reading Score')
plt.ylabel('Frequency')
plt.savefig('/content/students_performance/results/readingScores_distribution.png')
plt.show()
```

```python
# Analysis 5: Pie Chart showing the distribution of Parental Education Level

parental_education_counts = studentPerformance_clean_data['parental level of education'].value_counts()
plt.figure(figsize=(5, 5))
plt.pie(parental_education_counts, labels=parental_education_counts.index, autopct='%1.1f%%', startangle=50)
plt.title('Parental Education Level Distribution')
plt.savefig('/content/students_performance/results/Parental_Education_distribution.png')
plt.show()
```
