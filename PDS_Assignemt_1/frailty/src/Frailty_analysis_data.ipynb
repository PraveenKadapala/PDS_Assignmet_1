```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# importing frailty data
frailty_raw_data = pd.read_csv('/content/frailty/data_raw/frailty_raw_data.csv')

#Updated the column names
frailty_raw_data.columns = ['Height(in)', 'Weight(lb)', 'Age', 'Grip_Strength(kg)', 'Frailty']

#Replaced the categorical values of Frailty column with numerical values
frailty_raw_data['Frailty'].replace(['Y', 'N'], [0,1], inplace=True)

frailty_clean_data = frailty_raw_data

# Plotting the relationship between GripStrength and Frailty
plt.figure(figsize=(10, 6))
sns.boxplot(x='Frailty', y='Grip_Strength(kg)', data=frailty_clean_data, hue='Frailty', dodge=False)
plt.title('Relationship between GripStrength and Frailty')
plt.xlabel('Frailty')
plt.ylabel('GripStrength')
plt.savefig('/content/frailty/results/relation_between_gripstrength_frailty.png')
plt.show()
```
