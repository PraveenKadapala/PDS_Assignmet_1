```python
import pandas as pd
import datetime

# importing data
train_data_clean = pd.read_csv('/content/train_data_clean.csv')
train_data_clean.head()
```

```python
# Current Year
current_year = datetime.datetime.now().year
```

```python
# Calculating and creating new Current_Age Column for the cars.
train_data_clean['Current_Age'] = current_year - train_data_clean['Year']
```

```python
# Saving the modified data to a new CSV file
train_data_clean.to_csv('/content/car_age_data.csv', index=False)
```
