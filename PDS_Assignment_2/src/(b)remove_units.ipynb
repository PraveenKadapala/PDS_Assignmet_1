```python
import pandas as pd

# importing data
train_data_clean = pd.read_csv('/content/train_data_clean.csv')
train_data_clean.head()
```

```python
# Removing units from the attributes 'Mileage', 'Engine', 'Power', New_Price

train_data_clean['Mileage'] = train_data_clean['Mileage'].str.replace('kmpl', '').str.replace('km/kg', '').astype(float)
train_data_clean['Engine'] = train_data_clean['Engine'].str.replace('CC', '').astype(float)
train_data_clean['Power'] = train_data_clean['Power'].str.replace('bhp', '').astype(float)
train_data_clean['New_Price'] = train_data_clean['New_Price'].str.replace('Lakh', '').str.replace('Cr', '').astype(float)
```

```python
train_data_clean.head()
```

```python
# Saving the modified data to a new CSV file
train_data_clean.to_csv('/content/removed_units_data.csv', index=False)
```
