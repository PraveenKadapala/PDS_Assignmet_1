```python
import pandas as pd

# importing data
train_data = pd.read_csv('/content/train.csv')
```

```python
train_data.head()
```

```python
# Dropping unnamed column
train_data.drop(train_data.columns[0], axis=1, inplace=True)
```

```python
train_data.head()
```

```python
# Finding missing values in each column
missing_values = train_data.isna().sum()
```

```python
print(missing_values)
```

```python
# Replacing missing values with the mode for the columns having categorical values.
# Replace missing values with mean,  for the columns having numerical values.
for column in missing_values.index:
    if missing_values[column] > 0:
        if train_data[column].dtype == 'object':
            mode = train_data[column].mode()[0]
            train_data[column].fillna(mode, inplace=True)
        else:
            mean = train_data[column].mean()
            train_data[column].fillna(mean, inplace=True)
```

```python
train_data.head()
```

```python
# Saving the cleaned data to a new CSV file
train_data.to_csv('/content/train_data_clean.csv', index=False)
```
