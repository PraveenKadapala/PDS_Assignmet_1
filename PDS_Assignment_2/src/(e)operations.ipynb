```python
import pandas as pd

# importing data
train_data_clean = pd.read_csv('/content/train_data_clean.csv')
train_data_clean.head()
```

```python
# Select operation
#selecting the specific columns
selected_data = train_data_clean[['Name', 'Location', 'Year', 'Kilometers_Driven', 'Owner_Type', 'Mileage',
                  'Engine', 'Power', 'Seats', 'Price']]
selected_data.head()
```

```python
#Filter operation
#filtering dataset with location 'Chennai' and fuel_type 'Petrol'
filtered_data = train_data_clean[(train_data_clean["Location"] == "Chennai") & (train_data_clean["Fuel_Type"] == "Petrol")]
filtered_data.head()
```

```python
# Rename operation
#Renaming columns Mileage, Engine, Power with Mileage(kmpl), Engine(CC), Power(bhp)
renamed_columns = train_data_clean.rename(columns={"Mileage": "Mileage(kmpl)", "Engine": "Engine(CC)", "Power": "Power(bhp)"})
renamed_columns.head()
```

```python
# Mutate operation (
# Adding new columns "Age" and "Kilometers_Per_Year"
current_year = 2024
train_data_clean['Age'] = current_year - train_data_clean['Year']
train_data_clean['Kilometers_Per_Year'] = train_data_clean['Kilometers_Driven'] / train_data_clean['Age']
train_data_clean.head()
```

```python
# Arrange operation
# Arranging or sorting data based on kilometers_driven
arranged_data = train_data_clean.sort_values(by="Kilometers_Driven")
arranged_data.head()
```

```python
# Group By operation
# Sum of Kilometers_Driven and Mean price for each location
summary_data = train_data_clean.groupby("Location").agg({"Kilometers_Driven": "sum","Price": "mean"})
summary_data.head()
```
